```python
from PIL import Image
import os

def yolo_formatter():
    """
    Converts data from coords-idc.txt to YOLO-format
    """
    with open('coords-idc.txt', 'r') as file:
        data = file.readlines()
        for line in data:
            line = line[:-1].split(',') #remove newline and convert to list
            line = [int(char) for char in line] #convert to int (from str)
            with Image.open(f'idc_complete_images/{line[0]}.jpeg') as image:
                W, H = image.size #get image resolution
            with open(f'dataset/labels/train/{line[0]}.txt', 'w') as file:
                #Calculates the information for the bounding boxes, all have a minimum of 0
                xcenter = max((line[1]+line[3])/(2*W),0)
                ycenter = max((line[2]+line[4])/(2*H),0)
                width = max((line[3]-line[1])/W,0)
                height = max((line[4]-line[2])/H,0)
                file.write(f'1 {xcenter} {ycenter} {width} {height}')

dir = os.listdir('dataset/labels/train/')

if len(dir) <= 2: #Only call yolo_formatter() if there are now labels in dataset/labels/train/
    print('No existing labels found, creating new labels...')
    yolo_formatter()
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
model.train(data='dataset.yaml',
            epochs=15,
            single_cls=True)
```
